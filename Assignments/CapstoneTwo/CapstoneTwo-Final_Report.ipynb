### Modeling

Three different machine learning models that I tested my dataset are: Random Forest Classifier, Linear Regression, and GradientBoostingClassifier.

All models are tested to predict the risk level which are converted to:
> 2 : High\
> 1 : Medium\
> 0 : Low

These risk levels show how likely a person is to get Lung Cancer based on some of the living conditions, habits, as well as the symptoms you are experiencing.
