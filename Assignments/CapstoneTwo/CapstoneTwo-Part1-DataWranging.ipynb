```python
# import modules

import pandas as pd
import numpy as np
```

```python
# import data from the csv file

#data = pd.read_csv('cancer_patient_data sets.csv')
data = pd.read_excel('cancer patient data sets.xlsx')
data.head()
```

```python
# findout the row# and column# of the dataset

data.shape
print("Dataset has", data.shape[0],"rows and", data.shape[1],"columns.")
```

```python
# print column # and names to investigate what kind of data is in this file

col_count = 1

for col in data.columns:
    print(col_count, ":", col)
    col_count += 1
```

```python
# check for missing value

data.isnull().sum()
```

```python
# checking data format

data.info()
```

```python
# dropping Patient ID as it is not required in further analysis

data2 = data.drop(['Patient Id'], axis = 1)
data2
```
